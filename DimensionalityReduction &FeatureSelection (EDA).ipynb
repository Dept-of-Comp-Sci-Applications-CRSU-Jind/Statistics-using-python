Splitting the dataset

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_your, test_y= train_test_split (x,y,train_size=90,random_state=0)
train_x
```

FEATURE SELECTION METHODS

1. Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector=VarianceThreshold  (threshold =5)
selector
```

```python
train_x.shape
```

```python
selector.fit (train_x)
```

```python
selector.get_support()
```

```python
train_x
```

2.Univariate Selection

```python
from sklearn.linear_model import LinearRegression 

model=LinearRegression ()
model.fit (x,y)

importance=model.coef_
importance
```

```python
for i,v in enumerate(importance):

  print ('Feature:%0d,Score:%.5f' % (i,v))
```

```python
import matplotlib.pyplot as plt
plt.bar([x for x in range (len (importance ))],height=500)
plt.show()
```

Dimensionality Reduction

1.PCA(principle component analysis )

This is used to reduce the number of dimensions. 
This is help in resize the visualisation of multidimensional data

```python
from sklearn import decomposition 

Pca=decomposition.PCA()
Pca.n_components=2
Pca_data =Pca.fit_transform(train_x)
Pca_data
```

```python
Pca.explained_variance_ratio_
```

```python
Pca.n_components_
```

2.TSNE

```python
from sklearn.manifold import TSNE

m=TSNE (learning_rate=5)
tsne_features=m.fit_transform(x)
tsne_features
```

```python
plt.bar([x for x in range (len(tsne_features))],height =100)
plt.show()
```
