Measurement of Central tedency

mean : Average of the values in the data

```python
import numpy as np
#Create the dataset
items= [10,20,40,50,60,75,80]

#calculations
mean=np.mean(items)
median=np.median(items)

from statistics import mode
mode= mode(items)


#print the results
print("mean:", mean)
print("median:", median)
print("mode:",mode)

```

Measurement of Dispersion

```python
#calculating the range
Range =np.max(items) - np.min(items)

#calculating the variance 
variance=np.var(items)

#calculating standard deviation
std_dev= np.std(items)

#print results
print("Range:",Range)
print("variance: ",variance)
print("std_dev :",std_dev)
```

Measurement of Shape

```python
#import the skew module and numpy
from scipy.stats import skew
 
#calculate skewness using the scipy.stats
skewness = skew(items)

from scipy.stats import kurtosis
kurto = kurtosis(items)

#print the skewness
print("skewness :",skewness)
print("kurtosis: ",kurto)
```

Multivariate distributive statistics

```python
import numpy as np  
import pandas as pd 
#create dataset 
df=pd.DataFrame ({'Peter' : [20,30,40],'Johan':[10,25,30]})
df
```

Correlation Coefficient

```python
Corr=np.corrcoef(df['Peter'],df['Johan'])[0][1]

print("Correlation :",Corr)
```

```python
Covar=df['Peter'].cov(df['Johan'])

print("covariance :",Covar)
```

Normal Distribution

Use of cdf (Cumulative Distribution Function ) means probability of all outcomes like greater than or equal to any given value

```python
import numpy as np
import pandas as pd 
#create dataset of two students x and y obtain marks
df=pd.DataFrame({'Amit' :[2,4,6,8,10,14,20,22] ,'Selin':[3,5,9,16,19,25,30,40]})
df
```

```python
mean = df.mean()
mean
```

```python
#calculate standard deviation 
sigma =df.std ()
sigma
```

```python
#plotting graph 
import seaborn as sns 
sns.displot (df,height =5,kind="kde",aspect=2.7)
```

Probability of obtaining marks more than 5

```python
from scipy import stats
1-stats.norm.cdf(5,loc=mean,scale=sigma)
```

Probability of obtaining marks between 0 and 15

```python
stats.norm.cdf (15,loc=mean,scale=sigma)-stats.norm.cdf (0,loc=mean,scale=sigma)
```

Probability of obtaining marks less than 8



```python

stats.norm.cdf (8,loc=mean,scale=sigma)
```
